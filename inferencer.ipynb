```python
from fastai.vision.all import *
from fastai.vision.widgets import *

learn_inf = load_learner('resnet34-10.pkl', cpu=True)

btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()

def on_data_change(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(256,256))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_upload.observe(on_data_change, names=['data'])

file = open("47thaifood.png", "rb")
image2 = file.read()
image_headline = widgets.Image(value=image2,format='png',width='500')

display(VBox([image_headline, widgets.Label('Select your Thai food.'), btn_upload, out_pl, lbl_pred]))
```
